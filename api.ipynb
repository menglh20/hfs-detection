```python
import requests
import json, base64

'''
人脸检测与关键点检测
'''

request_url = "https://aip.baidubce.com/rest/2.0/face/v1/landmark"
filename = 'images/6_30.png'
file = open(filename, 'rb')
image = file.read()
file.close() 
img_base64 = base64.b64encode(image)

request_url = request_url + "?access_token=" + access_token
headers = {'content-type': 'application/json'}

req_data = json.dumps({'image': img_base64.decode(), 'image_type': 'BASE64', 'face_field': 'landmark72'})
response = requests.post(request_url, data=req_data, headers=headers)
if response:
    response = response.json()
    print(response)

result = response['result']
face = result['face_list'][0]

location = face['location']
print(location)

probability = face['face_probability']
print(probability)

landmark72 = face['landmark72']
for landmark in landmark72:
    print(landmark)
```

```python
import cv2

img = cv2.imread(filename)

# face
for landmark in landmark72[0:13]:
    cv2.circle(img, (int(landmark['x']), int(landmark['y'])), 1, (0, 0, 255), 2)

# left eye
for landmark in landmark72[13:22]:
    cv2.circle(img, (int(landmark['x']), int(landmark['y'])), 1, (0, 0, 255), 2)

# left eyebrow
for landmark in landmark72[22:30]:
    cv2.circle(img, (int(landmark['x']), int(landmark['y'])), 1, (0, 0, 255), 2)

# right eye
for landmark in landmark72[30:39]:
    cv2.circle(img, (int(landmark['x']), int(landmark['y'])), 1, (0, 0, 255), 2)

# right eyebrow
for landmark in landmark72[39:47]:
    cv2.circle(img, (int(landmark['x']), int(landmark['y'])), 1, (0, 0, 255), 2)

# nose
for landmark in landmark72[47:58]:
    cv2.circle(img, (int(landmark['x']), int(landmark['y'])), 1, (0, 0, 255), 2)

# mouth
for landmark in landmark72[58:72]:
    cv2.circle(img, (int(landmark['x']), int(landmark['y'])), 1, (0, 0, 255), 2)

# save image
cv2.imwrite('landmark.png', img)
```
